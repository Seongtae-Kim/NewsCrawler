```python
from collections import defaultdict
import os

art_list = [file for file in os.listdir("./output")
           if os.path.isfile(os.path.join("./output", file))]
#sum(1 for _ in art_list) # generator length
```

```python
len(art_list)
```

```python
dateDict = defaultdict()
for article in art_list:
    date = article[:8]
    if date in dateDict:
        dateDict[date].append(article)
    else:
        dateDict[date] = [article]
```

```python
dateDict
```

```python
for date in dateDict:
    date # folder name
    files = dateDict[date] # list of files
    if not os.path.exists("./output/"+date):
        os.mkdir("./output/"+ date)
        for file in files:
            os.replace("./output/" + file, "./output/"+date+"/"+file)
    else:
        for file in files:
            os.replace("./output/" + file, "./output/"+date+"/"+file)
```
